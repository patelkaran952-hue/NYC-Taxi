```python
import requests

url = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
response = requests.get(url)

with open("taxi_zone_lookup.csv", "wb") as f:
    f.write(response.content)
```

```python
import pandas as pd

taxi_lookup = pd.read_csv("taxi_zone_lookup.csv")
print(taxi_lookup.head())
```

## Creating a copy of raw data

```python
nyc_taxi = df_raw.copy()
```

```python
nyc_taxi.head()
```

```python
nyc_taxi.info()
```

## Replacing null values withy appropriate values

```python
nyc_taxi.isnull().sum()
```

```python
nyc_taxi.columns = nyc_taxi.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
nyc_taxi['passenger_count'] = nyc_taxi['passenger_count'].fillna(1)
nyc_taxi['ratecodeid'] = nyc_taxi['ratecodeid'].fillna(1)
```

```python
nyc_taxi['store_and_fwd_flag'] = nyc_taxi['store_and_fwd_flag'].fillna('N')
nyc_taxi['airport_fee'] = nyc_taxi['airport_fee'].fillna(0)
nyc_taxi['congestion_surcharge'] = nyc_taxi['congestion_surcharge'].fillna(0)
```

## Giving appropriate datatype to columns

```python
nyc_taxi['passenger_count'] = nyc_taxi['passenger_count'].astype(int)
```

## Creating new useful features and standardising data

```python
nyc_taxi['trip_duration'] = (nyc_taxi['tpep_dropoff_datetime'] - nyc_taxi['tpep_pickup_datetime']).dt.total_seconds() / 60
```

```python
nyc_taxi['pickup_hour'] = nyc_taxi['tpep_pickup_datetime'].dt.hour
```

```python
nyc_taxi['pickup_day'] = nyc_taxi['tpep_pickup_datetime'].dt.day_name()
```

```python
nyc_taxi['speed'] = nyc_taxi['trip_distance'] / (nyc_taxi['trip_duration'] / 60)
```

```python
nyc_taxi = nyc_taxi[(nyc_taxi['trip_duration'] > 0) & (nyc_taxi['trip_duration'] < 300)]
```

```python
nyc_taxi = nyc_taxi[(nyc_taxi['trip_distance'] > 0) & (nyc_taxi['trip_distance'] < 100)]
```

```python
nyc_taxi = nyc_taxi[(nyc_taxi['speed'] > 2) & (nyc_taxi['speed'] < 80)]
```

```python
nyc_taxi['fare_amount'].quantile([0.99])
```

```python
nyc_taxi = nyc_taxi[(nyc_taxi['fare_amount'] > 0) & (nyc_taxi['fare_amount'] < 100)]
```

```python
nyc_taxi = nyc_taxi[(nyc_taxi['passenger_count'] > 0) & (nyc_taxi['passenger_count'] <= 6)]
```

```python
nyc_taxi = nyc_taxi[(nyc_taxi['total_amount'] > 0)]
```

## Merging with lookup table

```python
taxi_lookup['Borough'] = taxi_lookup['Borough'].fillna('Unknown')
taxi_lookup['Zone'] = taxi_lookup['Zone'].fillna('Unknown')
taxi_lookup['service_zone'] = taxi_lookup['service_zone'].fillna('Unknown')
```

```python
nyc_taxi = nyc_taxi.merge(taxi_lookup, left_on='pulocationid', right_on='LocationID', how='left')
nyc_taxi = nyc_taxi.rename(columns={'zone': 'pickup_zone', 'borough': 'pickup_borough'})

nyc_taxi = nyc_taxi.merge(taxi_lookup, left_on='dolocationid', right_on='LocationID', how='left')
nyc_taxi = nyc_taxi.rename(columns={'zone': 'dropoff_zone', 'borough': 'dropoff_borough'})
```
